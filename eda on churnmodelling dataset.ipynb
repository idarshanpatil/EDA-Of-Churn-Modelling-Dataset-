```python
# Calculate the percentage of missing values for each column
missing = pd.DataFrame((b.isnull().sum() * 100) / b.shape[0]).reset_index()
missing.columns = ['column_name', 'missing_percentage']  # Renaming columns for clarity

# Plotting the percentage of missing values
plt.figure(figsize=(16, 5))
sns.pointplot(x='column_name', y='missing_percentage', data=missing)
plt.xticks(rotation=90, fontsize=7)
plt.title('Percentage of Missing Values')
plt.ylabel('Percentage')
plt.xlabel('Column Names')
plt.show()
```
